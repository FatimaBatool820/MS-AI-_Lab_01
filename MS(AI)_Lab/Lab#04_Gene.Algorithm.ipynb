```python
import numpy as np
import random

def create_8_queens_board():
    #Creates a random 8x8 chessboard with 8 queens placed randomly.
    board = np.full((8, 8), '.')
    for row in range(8):
        col = random.randint(0, 7)
        board[row, col] = 'Q'
    return board

def board_chromosome(board):
    #Extracts the chromosome (queen positions) from the board.
    queens_positions = np.argwhere(board == 'Q')
    cols = queens_positions[:, 1]
    return cols

def calculate_fitness_numpy(board):
    #Calculates the fitness
    queens_positions = np.argwhere(board == 'Q')
    num_queens = len(queens_positions)
    rows = queens_positions[:, 0]
    cols = queens_positions[:, 1]

    # Calculate column attacks
    unique_cols, counts = np.unique(cols, return_counts=True)
    unique_only = unique_cols[counts == 1]
    col_attacks = num_queens - len(unique_only)

    # Calculate diagonal attacks
    diagonal_attacks = 0
    for i in range(num_queens):
        for j in range(i + 1, num_queens):
            if abs(rows[i] - rows[j]) == abs(cols[i] - cols[j]):
                diagonal_attacks += 1

    total_attacks = col_attacks + diagonal_attacks
    fitness = num_queens - total_attacks  # Fitness is the number of non-attacking queens
    return fitness

def top_two_chromosomes(boards, chromosomes, fitness_values):
    #Selects the top two chromosomes based on fitness.
    combined = list(zip(fitness_values, boards, chromosomes))
    sorted_combined = sorted(combined, key=lambda x: x[0])
    top_two = sorted_combined[-2:]  # Get the two with highest fitness
    return top_two

def mutate_chromosome(chromosome):
    #Mutates a chromosome randomly
    index_to_mutate = random.randint(0, 7)
    new_gene_value = random.randint(0, 7)
    chromosome[index_to_mutate] = new_gene_value
    return chromosome

def single_point_crossover(parent1, parent2):
    #Performs single-point crossover
    crossover_point = 4  # You can make this random if needed
    offspring1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
    offspring2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
    return offspring1, offspring2

# Get population size from user
population_size = int(input("Enter the population size: "))

# Create initial population
boards = [create_8_queens_board() for _ in range(population_size)]
chromosomes = [board_chromosome(board) for board in boards]
fitness_values = [calculate_fitness_numpy(board) for board in boards]

    # Genetic Algorithm loop
for generation in range(5):  # Run for 100 generations (example)
    # Selection
    top_two = top_two_chromosomes(boards, chromosomes, fitness_values)
    parent1 = top_two[-1][2]
    parent2 = top_two[-2][2]

    # Crossover and Mutation
    offspring1, offspring2 = single_point_crossover(parent1, parent2)
    offspring1 = mutate_chromosome(offspring1)
    offspring2 = mutate_chromosome(offspring2)

    # Replace least fit individuals with offspring (example replacement strategy)
    # You might need a more sophisticated replacement strategy here
    boards[-2:] = [create_8_queens_board(),create_8_queens_board()] # Replace with the list of the boards
    chromosomes[-2:] = [offspring1, offspring2] # Replace with chromosomes
    fitness_values[-2:] = [calculate_fitness_numpy(boards[-2]), calculate_fitness_numpy(boards[-1])]  #Recalculate the fitness

    print(f"Generation {generation + 1}:")
    for i in range(population_size):
        print(f"Chromosome {i + 1}: {chromosomes[i]}, Fitness: {fitness_values[i]}")

    print("-" * 5)  # Separator between generations
```
