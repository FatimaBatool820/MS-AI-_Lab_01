# **Feature Scaling**

```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
heart=pd.read_csv('Heart_Disease.csv')
```

```python
heart.columns
```

```python
heart = heart.drop(columns=['State','LastCheckupTime','RemovedTeeth', 'HadAngina','HadCOPD','HadDepressiveDisorder',  'HadArthritis','DeafOrHardOfHearing', 'BlindOrVisionDifficulty',      'DifficultyConcentrating',
       'DifficultyDressingBathing', 'DifficultyErrands','ECigaretteUsage', 'ChestScan','HeightInMeters', 'WeightInKilograms','HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'TetanusLast10Tdap','HighRiskLastYear', 'CovidPos'], axis=1)
```

```python
heart.columns
```

```python
heart.shape
```

```python
heart.isnull().sum()
```

```python
heart.dtypes
```

```python
# Fill NaN values in numerical columns with the mean
numerical_cols = heart.select_dtypes(include=['number']).columns
heart[numerical_cols] = heart[numerical_cols].fillna(heart[numerical_cols].mean())

# Fill NaN values in categorical columns with the mode
categorical_cols = heart.select_dtypes(include=['object']).columns
heart[categorical_cols] = heart[categorical_cols].fillna(heart[categorical_cols].mode().iloc[0]) #to select first value in case many mode values
```

```python
heart.isnull().sum()
```

```python
#from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

numerical_cols = heart.select_dtypes(include=['number']).columns
heart[numerical_cols] = scaler.fit_transform(heart[numerical_cols])
print(type(heart[numerical_cols]))
```

```python
scaled_df=pd.DataFrame(heart[numerical_cols], columns=numerical_cols)
print(scaled_df)
```

# **Column Transformer**

```python
data=pd.DataFrame({
    'Age':[24,22,None,23,35],
    'sex':['female','female','female','male',None],
    'GeneralHealth':['Good','Fair','Excellent','Poor','Very good'],
    'BMI':[1.6,11.1,15.5,24.6,34.5]
})
```

```python
print("Original DataFrame:")
print(data)
```

```python
#Column Transformer
preprocessor = ColumnTransformer(transformers=[('age_imputer',SimpleImputer(strategy='mean'),['Age']),('ohe',OneHotEncoder(),['sex','GeneralHealth'])],remainder='passthrough')
transformed_data = preprocessor.fit_transform(data)
```

```python
print("\nTransformed Data Shape:")
print(transformed_data)
print(transformed_data.shape)
```

```python
transformed_df = pd.DataFrame(transformed_data)
print("\nTransformed DataFrame:")
print(transformed_df)
```

# **Function Tranformer**

```python
data=pd.DataFrame({
    'Age':[24,22,None,23,35],
    'sex':['female','female','female','male','male'],
    'GeneralHealth':['Good','Fair',None,'Poor','Good'],
    'BMI':[1.6,11.1,15.5,24.6,34.5]
})
print("Original DataFrame:")
print(data)
```

```python
def impute_generalhealth(X):
    X['GeneralHealth'] = X['GeneralHealth'].fillna(X['GeneralHealth'].mode()[0])  # Fill missing values
    print(X['GeneralHealth'])
    return X
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('age_imputer', SimpleImputer(strategy='mean'), ['Age']),
    ('generalhealth_imputer', FunctionTransformer(impute_generalhealth), ['GeneralHealth']),
    ('ohe', OneHotEncoder(), ['sex','GeneralHealth'])
], remainder='passthrough')
transformed_data = preprocessor.fit_transform(data)
```

```python
print("\nTransformed Data Shape:")
print(transformed_data.shape)
print(transformed_data)
```

```python
transformed_df = pd.DataFrame(transformed_data)
print("\nTransformed DataFrame:")
print(transformed_df)
```

# **Sklearn Pipeling**

```python
# Sample DataFrame
data=pd.DataFrame({
    'Age':[24,22,None,23,35],
    'sex':['female','female','female','male','male'],
    'GeneralHealth':['Good','Fair',None,'Poor','Good'],
    'BMI':[1.6,11.1,15.5,24.6,34.5]
})

print("Original DataFrame:")
print(data)

# Custom function to impute 'Embarked'
def impute_generalhealth(X):
    X['GeneralHealth'] = X['GeneralHealth'].fillna(X['GeneralHealth'].mode()[0])  # Fill missing values
    print (X['GeneralHealth'])
    return X
```

```python
#Column Transformer
preprocessor = ColumnTransformer(transformers=[
    ('age_imputer', SimpleImputer(strategy='mean'), ['Age']),
    ('GeneralHealth_encoder', Pipeline(steps=[
        ('imputer', FunctionTransformer(impute_generalhealth)),  # Impute generalhealth first
        ('onehot', OneHotEncoder())  # Then apply OneHotEncoder
    ]), ['GeneralHealth']),
    ('ohe', OneHotEncoder(), ['sex'])
], remainder='passthrough')  # Pass through other columns like 'BMI'
```

```python
# Transform the data
transformed_data = preprocessor.fit_transform(data)

# Output shape and transformed data
print("\nTransformed Data Shape:")
print(transformed_data.shape)
print(transformed_data)

transformed_df = pd.DataFrame(transformed_data)
print("\nTransformed DataFrame:")
print(transformed_df)
print(transformed_df.shape)
```
