```python
# Create a contingency table (cross-tabulation) for Pclass and Survived
ct=pd.crosstab(heart['SmokerStatus'], heart['Sex'])

# Plot the heatmap
#plt.figure(figsize=(8, 6))
sns.heatmap(ct,annot=True,fmt='d')
plt.title('Heatmap of Sex vs SmokerStatus')
plt.xlabel('Sex')
plt.ylabel('SmokerStatus')
plt.show()
```

```python
# Create a scatterplot for Age vs Fare
#plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='HadDiabetes', data=heart)

# Add titles and alabels
plt.title('Scatterplot of HadDiabetes vs BMI')
plt.xlabel('BMI')
plt.ylabel('HadDiabetes')

# Show the plot
plt.show()
```

```python
# Select only numerical columns for correlation analysis
numerical_features = heart.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numerical_features.corr()

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')

# Add titles and labels
plt.title('Correlation Matrix of Heart Disease Dataset Features')
plt.show()
```

# **Data Imputation**

```python
heart.isnull().sum()
```

```python
heart.isnull()
```

```python
sns.heatmap(heart.isnull(),cmap='viridis')
```

```python
# Fill NaN values in numerical columns with the mean
numerical_cols = heart.select_dtypes(include=['number']).columns
heart[numerical_cols] = heart[numerical_cols].fillna(heart[numerical_cols].mean())

# Fill NaN values in categorical columns with the mode
categorical_cols = heart.select_dtypes(include=['object']).columns
heart[categorical_cols] = heart[categorical_cols].fillna(heart[categorical_cols].mode().iloc[0]) #to select first value in case many mode values
```

```python
sns.heatmap(heart.isnull(),cmap='viridis')
```
